```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#using pandas to import your excel sheet as a dataframe
filename = 'records1.xml.xlsx'
df = pd.read_excel(filename)
```

```python
#df.head() gives the first 5 rows of the dataframe or you can pass an argument asking how many rows you want
df.head(5)
```

```python
# this is your excel sheet
df[:5]
```

```python
#sort by who paid the highest money
sorted_by_paid = df.sort_values(['PAID'], ascending=False)
sorted_by_paid
```

```python
#df.describe() gives all the info about the numerical values
df.describe()
```

```python
#you can also print the statistical information about each column, df.describe() gives you about all columns
print(df['PAID'].mean())
print(df['CHARGE'].mean())
```

```python
df['REMAINING'] = df['CHARGE'] - df['PAID']
```

```python
df
```

```python
#using pivot tables
df_subset = df[['NAME','PAID']]
df_subset.head()
```

```python
company_by_year = df_subset.pivot_table(index=['NAME'])
```

```python
#this gives the total amount paid by the company
company_by_year.head()
```

```python
company_by_year['PAID'].plot(kind="barh")
plt.show()
```

```python
df_sub1 = df[['NAME','PAID','REMAINING','CHARGE']]
df_sub1
```

```python
df_by_name = df_sub1.pivot_table(index=['NAME'])
df_by_name   
```

```python
fig = plt.figure()

plt.subplot(2, 2, 1)
plt.title('By Paid')
df_by_name['PAID'].plot(kind="bar")
plt.subplot(2, 2, 2)
plt.title('By Charge')
df_by_name['CHARGE'].plot(kind="bar")
plt.subplot(2, 2, 3)
plt.title('By Remaining')
df_by_name['REMAINING'].plot(kind="bar")

fig.subplots_adjust(wspace=1)
fig.subplots_adjust(hspace=1)
plt.show()
```
